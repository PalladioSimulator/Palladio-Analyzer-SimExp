```python
## Strategie 1
import re
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

pattern = r"Gen (\d+)"
avg_pattern = r"RT average:\s+(\d+\.\d+)"

def matches_pattern(string):
    return re.match(pattern, string)

def matches_avg(string):
    return re.match(avg_pattern, string)

def calc_means(data):
    grouped_values = {}
    
    for x, y in data:
        if x in grouped_values:
            grouped_values[x].append(y)
        else:
            grouped_values[x] = [y]

    result = [(x, sum(y_values) / len(y_values)) for x, y_values in grouped_values.items()]
    return result

def bla(file):
    with open(file, "r") as file:
        line = file.readline()
        currentGen = -1
        array = []
        list = []
        while line:
            gen_match = matches_pattern(line)
            avg_match = matches_avg(line)
            row = []
            if (bool(gen_match)):
                currentGen = gen_match.group(1)
                print("This is gen: " + str(currentGen))
                array.append(row)
                row = []
            elif (bool(avg_match)):
                #print("Found avg: " + str(avg_match.group(1)))
                row.append(avg_match.group(1)) 
                list.append((currentGen, float(avg_match.group(1))))
            line = file.readline()
        return list
    
               

    
list = bla("data/data_1_martina_10_100.txt")
listNeg = bla("data/data_1_seeded_negative_2.txt")
print(listNeg)
mean_tupels = calc_means(list)
neg_mean_tupels = calc_means(listNeg)

x, values = zip(*list)
y, mean_values = zip(*mean_tupels)
x2, neg_values = zip(*listNeg)
y2, neg_mean_values = zip(*neg_mean_tupels)

plt.scatter(x, values, color="blue", marker="o", label="individual values")
plt.scatter(y, mean_values, color="red", marker="x", label="mean")
#plt.scatter(x2, neg_values, color="green", marker="s", label="individual values2")
#plt.scatter(y2, neg_mean_values, color="orange", marker="*", label="mean2")
plt.xlabel("Generation")
plt.ylabel("RT average")
plt.title("2D Plot")
plt.grid()
plt.show()
```
